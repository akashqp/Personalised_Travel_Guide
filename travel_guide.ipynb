```python
import pandas as pd
import warnings
warnings.filterwarnings('ignore')
from sklearn.preprocessing import LabelEncoder, MinMaxScaler
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import folium
from geopy.geocoders import Nominatim

from sklearn.metrics.pairwise import cosine_similarity
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
df = pd.read_csv("Top Indian Places to Visit.csv", index_col=0)
df.head()
```

**Feature Engineering**:
   - Created metrics like `Cost-Effectiveness` and `Popularity Score` to improve ranking.
   - Categorized `Visit Duration` to allow time-based filtering.

```python
df["Cost-Effectiveness"] = df["Google review rating"] / np.log1p(df["Entrance Fee in INR"] + 1)

df["Popularity Score"] = (df["Google review rating"]**2) * df["Number of google review in lakhs"]

df["Visit Duration Category"] = pd.cut(
    df["time needed to visit in hrs"],
    bins=[0, 2, 5, np.inf],
    labels=["Short", "Medium", "Long"]
)

df["Proximity to Airport"] = df["Airport with 50km Radius"].apply(lambda x: 1 if x.strip().lower() == "yes" else 0)

df.head()
```

**Preprocessing**:
   - Normalized numerical data for consistency.
   - Encoded categorical features for compatibility with the recommendation engine.

```python
df.describe()
```

```python
# Check for missing values
missing_values = df.isnull().sum()

data_types = df.dtypes

missing_values, data_types
```

```python
# Fill missing values in 'Weekly Off' with 'None'
df["Weekly Off"].fillna("None", inplace=True)

# Normalize numerical data
scaler = MinMaxScaler()
numerical_columns = ['time needed to visit in hrs', 'Google review rating', 'Entrance Fee in INR', 'Number of google review in lakhs', 'Cost-Effectiveness', 'Popularity Score', 'Proximity to Airport']
df[numerical_columns] = scaler.fit_transform(df[numerical_columns])

# Encode categorical data
encoder = LabelEncoder()
categorical_columns = ['Zone', 'State', 'City', 'Name', 'Type', 'Establishment Year', 'Airport with 50km Radius', 'Weekly Off', 'Significance', 'DSLR Allowed', 'Best Time to visit', 'Visit Duration Category']
# df[categorical_columns] = df[categorical_columns].apply(lambda col: LabelEncoder().fit_transform(col.astype(str)))

# for col in categorical_columns:
#     df[col] = encoder.fit_transform(df[col].astype(str))  # Convert to string before encoding
```

```python
df.head()
```

```python
 # Combine categorical features into a single "profile"
df["Combined Features"] = (
    df["Zone"].astype(str) + " " +
    df["State"].astype(str) + " " +
    df["City"].astype(str) + " " +
    df["Type"].astype(str) + " " +
    df["Significance"].astype(str)
)

# Vectorize the combined features using TF-IDF
vectorizer = TfidfVectorizer()
feature_matrix = vectorizer.fit_transform(df["Combined Features"])
```

### Scoring Weights
- **Similarity Score (70%)**: Gives the highest importance to user preferences.
- **Popularity Score (20%)**: Ensures widely appreciated destinations rank higher.
- **Cost-Effectiveness (10%)**: Highlights budget-friendly options without overemphasis.

```python
# User Input: A user wants 'North Zone', 'Historical places', and low 'Entrance Fee'
user_query = "Northern Historical"
user_vector = vectorizer.transform([user_query])

# Calculate similarity scores
similarity_scores = cosine_similarity(user_vector, feature_matrix)

# Rank recommendations
df["Similarity Score"] = similarity_scores.flatten()

recommendations = df.sort_values(by="Similarity Score", ascending=False)


recommendations["Final Score"] = (
    0.7 * recommendations["Similarity Score"] +
    0.2 * recommendations["Popularity Score"] +
    0.1 * recommendations["Cost-Effectiveness"]
)
recommendations = recommendations.sort_values(by="Final Score", ascending=False)

geolocator = Nominatim(user_agent="travel_guide")

recommendations = recommendations.head(10)
# Function to get coordinates with city and state
def get_coordinates(row):
    location_query = f"{row['Name']}, {row['City']}, {row['State']}"
    try:
        loc = geolocator.geocode(location_query)
        if loc:
            print(location_query, loc.latitude, loc.longitude)
            return loc.latitude, loc.longitude
        else:
            print(f"Coordinates not found for: {location_query}")
            return None, None
    except Exception as e:
        print(f"Error fetching coordinates for: {location_query}, {e}")
        return None, None

# Apply the function to get coordinates
recommendations["Latitude"], recommendations["Longitude"] = zip(
    *recommendations.apply(get_coordinates, axis=1)
)
# Display top recommendations
print("Top Recommendations:")
print(recommendations[["Name", "City", "Type", "Significance", "Similarity Score", "Final Score", "Popularity Score", "Cost-Effectiveness"]])
```

```python
# Plot Final Score Distribution
top_recommendations = recommendations.head(10)
sns.barplot(x="Final Score", y="Name", data=top_recommendations)
plt.title("Top 10 Recommendations")
plt.xlabel("Final Score")
plt.ylabel("Place Name")
plt.show()
```

```python
# Initialize map
map_center = [28.6139, 77.2090]  # Example: Delhi coordinates
travel_map = folium.Map(location=map_center, zoom_start=5)

# Add markers for recommendations
for _, row in top_recommendations.iterrows():
    if pd.isna(row["Latitude"]) or pd.isna(row["Longitude"]):
        continue
    print(row["Latitude"], row["Longitude"], row["Name"])
    folium.Marker(
        location=[row["Latitude"], row["Longitude"]],  # Add lat/lon data to dataset
        popup=f"{row['Name']} - Score: {row['Final Score']:.2f}"
    ).add_to(travel_map)

# Display map
travel_map
```
